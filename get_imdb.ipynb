```python
all_imdb=list(all_imdb)
movie_ids = []
for movie in all_imdb:
    movie_str = repr(movie)
    start_index = movie_str.find("id:") + 3
    end_index = movie_str.find("[http]", start_index)
    movie_id = movie_str[start_index:end_index]
    movie_ids.append(movie_id)
```
