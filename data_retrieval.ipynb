```python
mismatch_v=[]
mismatch_tr=[]
for y_1,y_2 in zip(years,movies.iloc[:]["Movie release date"].str[:4]):
    if str(y_1)!=str(y_2) and len(str(y_2))>3:
        mismatch_v.append((y_1,y_2))
        try:
            if np.abs(int(y_1)-int(y_2))>1:
                mismatch_tr.append((y_1,y_2))
        except:
            continue
```
